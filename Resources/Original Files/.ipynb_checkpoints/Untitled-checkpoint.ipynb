```python
# Set up dependencies
import pandas as pd
import os as os
```

```python
# Import the raw data into a DataFrame: Average Income

csvpath = ("11100190_Annual Average Income_Nat_Prov_City.csv")
Average_Income = pd.read_csv(csvpath)
Average_Income.head()
```

```python
# Selecting final columns

Average_Income_clean = Average_Income[["REF_DATE","GEO","Income concept", "Economic family type", "VALUE"]]
Average_Income_clean.head()
```

```python
# Renaming column labels

Average_Income_clean = Average_Income_clean.rename(columns={"REF_DATE": "Date (Year)", "GEO": "Location", 
                                                            "Income Concept": "Type of Income", 
                                                            "Economic family type": "Group", "VALUE": "Average Amount(C$)"})

Average_Income_clean.head()
```

```python
Average_Income_clean["Date (Year)"].max()
```

```python
Average_Income_clean["Date (Year)"].min()
```

```python
Average_Income_clean["Location"].unique()
```

```python
Average_Income_clean["Income concept"].unique()
```

```python
Average_Income_clean["Group"].unique()
```

```python
Average_Income_clean.describe()
```

```python
# Saving final dataframe in the resources folder

Average_Income_clean.to_csv("../Average_Income Canada_1976_2008.csv",index=False)
```
